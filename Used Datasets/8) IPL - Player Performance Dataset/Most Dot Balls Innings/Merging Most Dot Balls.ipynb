```python
import pandas as pd
import os
```

```python
years = range(2008, 2023)
files = [fr"C:\Users\Mohak kansal\Documents\Coding\Projects\IPL Auction Analytics & Team Performance Prediction Dashboard\Data Collection and Cleaning\Used Datasets\8) IPL - Player Performance Dataset\Most Dot Balls Innings\Most Dot Balls Innings - {year}.csv" for year in years]

all_dfs = []

for file in files:
    df = pd.read_csv(file)

    year = file.split(' - ')[-1].split('.')[0]
        
    df['Year'] = year

    all_dfs.append(df)

merged_df = pd.concat(all_dfs, ignore_index=True)

merged_df.head()
```

```python
merged_df['Maid'] = merged_df['Maid'].fillna(0)

merged_df['Maid'] = merged_df['Maid'].astype(int)
```
