### Data Cleaning

#### Handling null values

```python
# Fill expected nulls with 'None'
for col in ['player_dismissed', 'dismissal_kind', 'fielder', 'extras_type']:
    df[col].fillna('None', inplace=True)

# Fill numeric nulls with 0
for col in ['batsman_runs', 'extra_runs', 'total_runs']:
    df[col].fillna(0, inplace=True)
```

#### Standardize text

```python
# Strip whitespace and standardize case for team and player columns
for col in ['batting_team', 'bowling_team', 'batter', 'bowler', 'non_striker', 'player_dismissed', 'fielder']:
    df[col] = df[col].str.strip().str.title()  # or .str.upper() for all caps
```
